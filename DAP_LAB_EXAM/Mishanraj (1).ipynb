```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns



```

```python
df = pd.read_csv('COVID.csv')
df
```

```python
df.isnull().sum()
```

```python
df.groupby('Region')[['Confirmed','Recovered','Deaths']].sum()
```

```python
#Removing all the records where the confirmed cases is less than 10
df[df['Confirmed']>=10]
```

```python
#3) In which Region, maximum number of Confirmed cases were recorded ?
max_confirmed=df.groupby('Region').Confirmed.sum().sort_values(ascending = False).head(1)
```

```python
plt.plot(df['Region'], df['Confirmed'])  # Plotting 'y' against 'x'
plt.xlabel('Region')
plt.ylabel('Confirmed')
plt.title('Region VS Confirmed cases')
plt.grid(True)
plt.show()
```

```python
#Q. 4) In which Region, minimum number of Deaths cases were recorded ?
min_deaths=df.groupby('Region').Deaths.sum().sort_values(ascending=True).head(5)
min_deaths
```

```python
plt.plot(df['Region'], df['Deaths'])  # Plotting 'Deaths' against 'Region'
plt.xlabel('Region')
plt.ylabel('Deaths')
plt.title('Plot of Deaths by Region')
plt.grid(True)
plt.show()
```

```python
#Q. 5) How many Confirmed, Deaths & Recovered cases were reported from India till 29 April 
df[df['Region'] == 'India']
```

```python
#Q. 6-A ) Sort the entire data wrt No. of Confirmed cases in ascending order.
df.sort_values('Confirmed',ascending = True)
```

```python

#Q. 6-B ) Sort the entire data wrt No. of Recovered cases in descending order
df.sort_values('Recovered', ascending= False)
```
